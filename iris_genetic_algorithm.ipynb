```python
plt.title('Classification Count')
sns.countplot(iris_df['classification'])
```

```python
plt.figure(figsize=(17,9))
plt.title('Comparison between various species based on sapel length and width')
sns.scatterplot(iris_df['sepal_length'],iris_df['sepal_width'],hue =iris_df['classification'],s=60)
```

```python
plt.figure(figsize=(16,9))
plt.title('Comparison between various species based on petal lenght and width')
sns.scatterplot(iris_df['petal_length'], iris_df['petal_width'], hue = iris_df['classification'], s= 60)
```
